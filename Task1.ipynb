```python
data.reset_index(inplace=True, drop=True)
data
```

```python
data = data.dropna()
```

```python
data['payload_bytes_3'] = data['payload_bytes_3'].str.slice(0, 4)
```

```python
data['payload_bytes_3'] = data['payload_bytes_3'].map(lambda x: 1 if x in ['1603', '4544', '1403', 'd500'] else 0)
```

```python
data['dst_ip'] = data['dst_ip'].map(lambda x: 1 if x == '130.87.143.208' else 0)
```

```python
data
```

```python
X = data.drop(columns=['protocol'])
y = data['protocol']
```

```python
y = y.astype(int)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
models = [
    DecisionTreeClassifier(),
    GradientBoostingClassifier(), 
    RandomForestClassifier(),
    XGBClassifier(),
    CatBoostClassifier()
]
```

```python
def make_prediction(m, X_train, y_train, X_test, y_test):
    model = m
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    print(
        'Accuracy:{:.2f} Precision:{:.2f} Recall:{:.2f} F1:{:.2f} '.format(
            accuracy_score(y_test, y_pred),
            precision_score(y_test, y_pred, average='micro'),
            recall_score(y_test, y_pred, average='micro'),
            f1_score(y_test, y_pred, average='micro'),
        )
    )
```

```python
for i in models:
    print(i)
    make_prediction(
        m=i,
        X_train=X_train,
        y_train=y_train,
        X_test=X_test,
        y_test=y_test,
    )
```
